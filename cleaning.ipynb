```python



# Step 1: Fix the MRP column first by copying prices from matching duplicate rows
mrp_by_group = df.groupby(['Product_Name', 'Product_Line', 'Gender_Category'])['MRP'].transform('first')
df['MRP'] = df['MRP'].fillna(mrp_by_group)

# Handle any global fallback for prices just in case
df['MRP'] = df['MRP'].fillna(df['MRP'].median())

# Clean up discounts safely (fill missing discounts with 0.0)
discount_clean = df['Discount_Applied'].fillna(0.0).clip(0.0, 1.0)

# Step 2: Implement your two conditional rules row-by-row
def fix_math_rows(row, discount_series):
    units = row['Units_Sold']
    mrp = row['MRP']
    rev = row['Revenue']
    disc = discount_series[row.name]
    
    # Rule 2: If Revenue is already written, back-calculate missing Units_Sold
    if pd.notna(rev) and rev > 0 and pd.isna(units):
        calculated_units = rev / (mrp * (1 - disc))
        units = round(calculated_units)
    elif rev == 0 and pd.isna(units):
        units = 0.0
        
    # Rule 1: If Units and MRP exist, calculate/re-verify the Revenue
    if pd.notna(units) and pd.notna(mrp):
        rev = units * mrp * (1 - disc)
        
    return pd.Series([units, round(rev, 2)])

# Apply the rules directly to update your dataframe columns cleanly
df[['Units_Sold', 'Revenue']] = df.apply(lambda r: fix_math_rows(r, discount_clean), axis=1)

```

```python
df['Discount_Applied']=df['Discount_Applied'].fillna(0).clip(upper =  0.99)
```

```python
df['Profit']= df['Profit'].abs()
def apply_accurate_reverse_engineering(row):
    units = row['Units_Sold']
    rev = row['Revenue']
    mrp = row['MRP']
    profit = row['Profit']
    disc = row['Discount_Applied']
    
    is_rev_zero = pd.isna(rev) or rev == 0.0
    is_units_zero = pd.isna(units) or units == 0.0
    
    if is_rev_zero and is_units_zero:
        if pd.notna(profit) and profit != 0:
            # 1. Estimate base value from profit margin
            estimated_discounted_value = abs(profit) / 0.35
            
            # 2. Find the true price of one shoe after discount
            discounted_price_per_unit = mrp * (1 - disc)
            
            # 3. Solve for units cleanly (e.g., 2201.29 / 5125.66 = ~0.43 -> rounded to 1 unit)
            if discounted_price_per_unit > 0:
                units = max(1, round(estimated_discounted_value / discounted_price_per_unit))
            else:
                units = 1
            
            # 4. Final Revenue matches the physical unit sale!
            rev = round(units * mrp * (1 - disc), 2)
        else:
            rev = 0.0
            units = 0
            
    return pd.Series([int(units), rev])

# Overwrite only the target rows
df[['Units_Sold', 'Revenue']] = df.apply(apply_accurate_reverse_engineering, axis=1)
```
